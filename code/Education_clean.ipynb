In addition to the other predictor variables under consideration, we aim to examine the impact of full-time teachers—or the full-time equivalent of part-time teachers—on ACT scores. To accomplish this, we will import additional data and ensure it is properly cleaned so that it can be temporarily merged with the edgap dataframe for quick exploratory analysis to determine whether the dataset we have is usable for our exploratory analysis.

Note: An issue was encountered because the dataset appears as a single column; this occurs because the NCES site has embedded metadata in the first row. To resolve this, we need to skip the first row when importing the data, and at the same time, we can assign more descriptive column names.

```python
#Load CSV with correct header and delimiter
full_time_teachers_count_info = pd.read_csv(
    "ELSI_csv_export_6389617089029104698184.csv",
    skiprows=6,          # skip metadata above header
    encoding="utf-8-sig",
    delimiter=',',       # ensure proper column separation
    on_bad_lines="skip"  # skip junk footer lines
)

#Clean the column names
full_time_teachers_count_info.columns = (
    full_time_teachers_count_info.columns
    .str.strip()
    .str.replace('\ufeff', '', regex=True)
)

#Rename the main columns to give better descriptive name
full_time_teachers_count_info.rename(columns={
    full_time_teachers_count_info.columns[0]: 'school_name',
    full_time_teachers_count_info.columns[1]: 'state',
    full_time_teachers_count_info.columns[2]: 'fte_teachers',
    full_time_teachers_count_info.columns[3]: 'id'
}, inplace=True)

# Drop footer rows where 'state' is missing
full_time_teachers_count_info = full_time_teachers_count_info.dropna(subset=['state'])


```

It was noted that fte_teachers variable in the full_time_teachers_count_info dataframe is of object type 'object'. To cast it to float to ensure that it's datatype are consistent with other predictor variables:

```python
full_time_teachers_count_info = full_time_teachers_count_info.copy()

full_time_teachers_count_info['fte_teachers'] = (
    full_time_teachers_count_info['fte_teachers']
    .astype(str)
    .replace(['–', '-', 'NA', ''], np.nan)
    .str.replace(',', '', regex=False)
)

full_time_teachers_count_info['fte_teachers'] = pd.to_numeric(
    full_time_teachers_count_info['fte_teachers'], errors='coerce'
)

print(full_time_teachers_count_info['fte_teachers'].dtype)  # float64
print(full_time_teachers_count_info['fte_teachers'].isna().sum())  # number of NaNs
```

Note: If we drop rows that have NaNs for any of our socioeconomic variables, then this will negatively affect our analysis using the variable where data was present; hence we will not drop the rows in this dataset that are missing socioeconomical variables. We will instead impute the missing values.

In order to handle the missing values, we will have to impute the missing values for the socioenomical factors that are omitted. We will be replacing with missing values that are obtained through prediction from the data. We will be using an iterativeImputer to do this.
The basic idea of an IterativeImputer is that instead of of just replacing missing values with a fixed statistic (mean, median, mode), the Iterative Imputer predicts missing values using other features. It will be using regression analysis to look at the predictors so if there is a missing value of one of them, it can use the best guess of that value based on its relationship to the other variables. 

To define our predictor variables:

```python
#Define predictor variables 
predictor_vars = [
    'percent_college', 'rate_unemployment', 'percent_married', 'median_income', 'percent_lunch', 'fte_teachers'
]
```

```python
#import Iterative imputer:
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer
```

```python
#define an iterative imputer object
imputer = IterativeImputer()
```

To fit the imputer with numerical values that we have:

```python
numerical_predictors = df[predictor_vars].select_dtypes(include='number').columns.to_list()
print(numerical_predictors)
```

The numerical_predictors have been outputted.

To fit the imputter with data from the specific columns:

```python
imputer.fit(df.loc[:, numerical_predictors])
```

Next, apply transform method on the columns to actually replace values. This will impute missing values in training data:

```python
df.loc[:, numerical_predictors] = imputer.transform(df.loc[:, numerical_predictors])
```

To confirm that values have been successfully replaced:

```python
df.isna().sum().to_frame(name='Number of Missing Values')  #show total computation of missing values for each column
```

Our NaNs have been successfully replaced with the imputed data as there are no missing data. The clean data can now be exported for further analysis.
