To perform quality control to look for values that are outside the range of possible values for a particular variable:

Note: ACT score should be between 1 and 36 and we have multiple values that correspond to a percentage or a proportion-should be between 0 and 1.

To determine our max and min variables to understand the range of our values:

```python
df.describe(include='number')
```

From the descriptive statistics, we can see that we have values that are outisde the range of possible values.
Our ACT score is between 1 and 36, but we can see the minium value for average_act(average ACT score) is -3.070818. It shouldn't be possible to have a negative score, hence we have to fix that. 
Also, the percentage_lunch(percentage of student eligible for reduced price lunch) is also negative but should be between 0 and 1.

```python
df.loc[df['average_act'] < 1, 'average_act'] = np.nan
```

Note: If we drop rows that have NaNs for any of our socioeconomic variables, then this will negativef=ly affect our analysis using the variable where data was present; hence we will not drop the rows in this dataset that are missing socioeconmical variables. We will instead impute the missing values.

In order to handle the missing values, we will have to impute the missing values for the socioenomical factors that are omitted. We will be replacing with missing values that are obtained through prediction from the data. We will be using an iterativeImputer to do this. This will be using regression analysis to look at the predictors so if there is a missing value of one of them, it can use the best guess of that value based on its relationship to the other variables. 

To define our predictor variables:

```python
#Define predictor variables 
predictor_vars = [
    'percent_college', 'rate_unemployment', 'percent_married', 'median_income', 'percent_lunch', 'fte_teachers'
]
```

```python
#import Iterative imputer:
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer
```

```python
#define an iterative imputer object
imputer = IterativeImputer()
```

To fit the imputer with numerical values that we have:

```python
numerical_predictors = df[predictor_vars].select_dtypes(include='number').columns.to_list()
print(numerical_predictors)
```

The numerical_predictors have been outputted.

To fit the imputter with data from the specific columns:

```python
imputer.fit(df.loc[:, numerical_predictors])
```

Next, apply transform method on the columns to actually replace values. This will impute missing values in training data:

```python
df.loc[:, numerical_predictors] = imputer.transform(df.loc[:, numerical_predictors])
```

To confirm that values have been successfully replaced:

```python
df.isna().sum().to_frame(name='Number of Missing Values')  #show total computation of missing values for each column
```

Our NaNs have been successfully replaced with the imputed data as there no missing data.
