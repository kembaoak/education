To add a regression lines to our plots and format the pair plot:


```python
fig = sns.pairplot(
    edgap.drop(columns = "NCESSCH School ID"),
    
    
    kind="reg", #indicates kind is a regression
    plot_kws = {
        "line_kws": {"color": "blue"},
        "scatter_kws": {"alpha":0.5, "color": "k", "s":7},
    },
)

for ax in fig.axes.flat:
    if ax.get_xlabel() == 'CT Median Household Income':
        ax.ticklabel_format(style='sci', axis='x', scilimits = (0,0)) #apply scientific notation
    ax.set_xlabel(ax.get_xlabel(), fontsize=8, rotation=30, ha='right') #x-axis label size and rotation
    ax.set_ylabel(ax.get_ylabel(), fontsize=8) #Y-axis label size
    
    #Rotate x-axis tick labels
    plt.setp(ax.get_xticklabels(), rotation=30, ha='right')
    
plt.show()

```

The output show a scatterplots with a regression line added. The regression line shows the best fit linear relationship between the two variables.

Join the dataframes and call the result df.
Note: left join allows us combine tables based on common key('id' in this case). It keeps all rows from the left table and adds matching rows from the right table. If there is no match, the result will have NaN for missing columns from the right table. 

```python
df = edgap.merge(
    school_information,
    how='left',
    on='id'
)
```

To impute NaN values and export the clean data set: 

To define the predictor variables to be rate_unemployment, percent_college, percent_married, median_income, percent_lunch, state, and charter we want to use to predict the ACT scores.

```python
predictor_variables = [
    'rate_unemployment',
    'percent_college',
    'percent_married',
    'median_income',
    'percent_lunch',
    'state',
    'charter'
]
```

To use the iterative imputer to replace missing values in the columns corresponding to predictor variables in the analysis.
We need to install and import scikit-learn package. 
Note: Scikit-learn is the library that lets you build, train, test, and evaluate machine learning models — all with a few lines of code.

```python
#To install sklearn package
!pip install scikit-learn
```

To import the required libraries:

```python
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer
```

```python
imputer = IterativeImputer()
```

To define the columns you want to use the imputation process:


```python
#defining numerical predictors
numerical_predictors = df[predictor_variables].select_dtypes(include='number').columns.to_list()
print(numerical_predictors)
```

The output above lists the numerical_predictors

To fit the imputer with data from the columns: 
This allows us to learn the relationships between those columns then apply the transform method the replace the values we would like to replace. 

```python
imputer.fit(df.loc[:, numerical_predictors])
```

To impute the missing values in the training data

```python
df.loc[:, numerical_predictors] = imputer.transform(df.loc[:, numerical_predictors])
```

To confirm that we have indeed replaced the missing values in the dataframe:

```python
df.isna().sum().to_frame(name='Number of Missing Values')  #show total computation of missing values for each column
```

The output shows that all missing values have been replaced
